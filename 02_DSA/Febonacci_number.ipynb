```python
def febonacci(n):
    
    a = 0
    b = 1
    
    # Handle 0 or negative number, user input
    if n <= 0 :
        return
    # Handle 1 number, user input
    elif n == 1 :
        print (a)
        return
    
    print (a, b, end=" ")
    
    #-2 coz a,b already print
    for i in range (n - 2):
        
        sum = a + b
        a = b
        b = sum 
        
        print (b, end=" ")
```
