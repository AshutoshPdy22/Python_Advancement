```python
def perfect_no (n):
        
    sum = 0
    #Handle expection of input  0 , 1
    if n == 0 or n == 1 :
        return False
    
    # main logic for divition with every number in sequence
    for i in range (1, n):
        
        if  n % i == 0 :
            # adding only the perfect number
            sum += i
    return sum == n
```

```python
if __name__ == "__main__":
    
    n = int(input("ender number for checking Perfect number"))
    if perfect_no (n):
        print (f"Yes {n} is perfect number")
    else :
        print (f"No {n} is not perfect number")
```
