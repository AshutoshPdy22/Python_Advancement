## Using Tuple Unpacking

```python
x, y = 10, 50

x, y = y, x  # swapping
print("x:", x)
print("y:", y)
```

## Using XOR Operator (just learn the code)

-  swap two variables at the bit level example = x as a 1100

```python
x, y = 12, 50

x = x ^ y  
y = x ^ y  
x = x ^ y  

print("x:", x)
print("y:", y)
```
