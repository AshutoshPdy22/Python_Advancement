```python
def prime_number (n):
    
    # Handle 0, 1 and negative numbers Coz they not prime.
    if n <= 1:
        return False
    
    # Check can it diviable by number which came before it.
    for i in range (2, n):
        if n % i == 0 :
            return False
        
    # Yes it a prime number.
    return True
```

```python
if __name__ == "__main__":
    n = int (input("Enter a number to check Prime"))
    print (f"user checking {n} as a prime number")
    prime_number (n)
    
    if prime_number (n):
        print (f"yes {n} is a prime number")
    else:
        print (f"No {n} is not a prime number")
```
