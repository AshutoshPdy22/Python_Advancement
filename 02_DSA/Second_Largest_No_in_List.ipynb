```python
def find_2nd_large (a):
    if a[0]>a[1]:
        max1 = a[0]
        max2 = a[1]
    else :
        max1 = a[1]
        max2 = a[0]
    
    for i in range (2, len(a)):
        if a[i] > max1 :
            max2 = max1
            max1 = a[i]
            
        elif a[i] > max2:
            max2 = a[i]
    return max2
```
