```python
def palindrome (n) :
    
    s = str(n)
    length = len(s)
    
    for i in range (length // 2) :
        if s [i] != s[length - i - 1] :
            return False
    return True
```

```python
if __name__ == "__main__":
    n = 112211
    if palindrome(n):
        print ("yes it is")
    else :
        print ("no it is not")
```
