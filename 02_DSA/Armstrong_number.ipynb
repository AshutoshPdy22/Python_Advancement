```python
def armstrong (n):
    number = str (n)
    digits = len(number)
    output = 0
    
    for i in number :
        output += int(i) ** digits
    return output == n
```

```python
if __name__ == "__main__" :
    n = 153
    
    if armstrong (n):
        print ("yes it is")
    else :
        print ("no it is not")
    
```
