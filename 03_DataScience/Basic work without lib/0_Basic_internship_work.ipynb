### Step 3: Display users and their connections

```python
def display_user (data):
    
    print("Users and Their Connections:\n")
    for user in data["users"]:
        print(f"Mr. {user['name']} alloated id as {user['id']}, love to do {user['liked_pages']} & there friend is {user['friends']}")
        
    print("\nPages:\n")
    for page in data['pages']:
        print(f"Subject id is {page['id']} = {page['name']}")
```
