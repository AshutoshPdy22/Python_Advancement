```python
cols=["flength","fwidth","fsize","fconc","fconcl","fasym","fm3long","fm3trans","falpha","fdist","class"]
df=pd.read_csv("magic04.data",names=cols)
df.head()
```

so after that I notice that in "class label" we have g and h values. In this datasets
- g stands for Gammas
- h stand for hadrons

```python
df["class"].unique()
```

df["class"].unique() use => Python used with pandas (a library for data analysis)
- df: This is your DataFrame, a table-like structure containing your data.
- "class": This is the column in the DataFrame that you're interested in.
- .unique(): This function finds all the unique (distinct) values in that column.

Computer are smart in understanding numbers so we change this above command and set condition "if g is equal to class then print true (1), if not then print flase(0)"

```python
df ["class"] = (df["class"] == "g").astype(int)
```

above input features predict the labels as output of machince by:
1. df["class"] == "g":
This creates a boolean series (True/False) where each row is checked to see if the value in the class column is "g".
Example: If df["class"] = ["g", "h"], the result would be:
[True, False].
2. .astype(int):
This converts the boolean values (True/False) into integers (1/0).
True becomes 1, and False becomes 0.
3. df["class"] = ...:
The resulting integers (0 or 1) are assigned back to the class column in the DataFrame.

```python
for label in cols[:-1]:
  plt.hist(df[df["class"]==1] [label], color="blue",label="gamma", alpha=0.7, density=True)
  plt.hist(df[df["class"]==0] [label], color="red",label="hadron", alpha=0.7, density=True)
  plt.title(label)
  plt.ylabel("probability by me")
  plt.xlabel(label)
  plt.legend()
  plt.show()
```
